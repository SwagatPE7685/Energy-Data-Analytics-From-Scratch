> **1. Matplotlib**

> **2. Seaborn**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sw = np.linspace(0,1,100)
```

```python
sw
```

```python
#Krw = Relative permeability of Water
krw = sw**3
```

```python
krw
```

```python
# kro = Relative permeability of oil
kro = (1-sw)**2
```

```python
kro
```

```python
df_relperm = pd.DataFrame({'Sw': sw , 'krw' : krw , 'kro' :kro})
```

```python
df_relperm
```

```python
df_relperm.plot()
```

```python
df_relperm.plot(subplots = True)
```

```python
#Step 1 : plotting your krw vrs sw 

plt.plot(sw,krw)
```

```python
#Step 2: plotting kro vrs sw

plt.plot(kro,sw)
```

```python
#Step 3: plotting kro & krw

plt.plot(sw,krw);
plt.plot(sw,kro)
```

```python
#Step 4: give x and y axis label

plt.plot(sw,kro);
plt.plot(sw,krw)

plt.xlabel('Water saturation (Fraction)')
plt.ylabel('Relative permeability(fraction)')
```

```python
#Step 5: Grid

plt.plot(sw,kro);
plt.plot(sw,krw)

plt.xlabel('Water saturation (Fraction)')
plt.ylabel('Relative permeability(fraction)')

plt.grid(True)
```

```python
#step 6:  adding legend to the plot

plt.plot(sw,kro,label = 'Oil Relative permeability(kro)')
plt.plot(sw,krw,label = 'Water Relative permeability(krw)')

plt.xlabel('Water saturation (Fraction)')
plt.ylabel('Relative permeability(fraction)')

plt.grid(True)
plt.legend()
```

```python
#You can also give location of label
plt.plot(sw,kro,label = 'Oil Relative permeability(kro)')
plt.plot(sw,krw,label = 'Water Relative permeability(krw)')

plt.xlabel('Water saturation (Fraction)')
plt.ylabel('Relative permeability(fraction)')

plt.grid(True)
plt.legend(loc = 'upper left')
```

```python
plt.plot(sw,kro,label = 'Oil Relative permeability(kro)')
plt.plot(sw,krw,label = 'Water Relative permeability(krw)')

plt.xlabel('Water saturation (Fraction)')
plt.ylabel('Relative permeability(fraction)')

plt.grid(True)
plt.legend(loc = 'upper right')
```

```python
plt.plot(sw,kro,label = 'Oil Relative permeability(kro)')
plt.plot(sw,krw,label = 'Water Relative permeability(krw)')

plt.xlabel('Water saturation (Fraction)')
plt.ylabel('Relative permeability(fraction)')

plt.grid(True)
plt.legend(loc = 'best')
```

```python
#Step 7 change the size of plot

plt.figure(figsize = (16,10))
plt.plot(sw,kro,label = 'Oil Relative permeability(kro)')
plt.plot(sw,krw,label = 'Water Relative permeability(krw)')

plt.xlabel('Water saturation (Fraction)')
plt.ylabel('Relative permeability(fraction)')

plt.grid(True)
plt.legend(loc = 'best')
```

```python
#Step 8 : change the lines color

plt.figure(figsize = (16,10))
plt.plot(sw,kro,label = 'Oil Relative permeability(kro)',color = 'green')
plt.plot(sw,krw,label = 'Water Relative permeability(krw)',color = 'pink')

plt.xlabel('Water saturation (Fraction)')
plt.ylabel('Relative permeability(fraction)')

plt.grid(True)
plt.legend(loc = 'best')
```

```python
#Step 9: changing the line thickness

plt.figure(figsize = (16,10))
plt.plot(sw,kro,label = 'Oil Relative permeability(kro)',color = 'green',lw = 4)
plt.plot(sw,krw,label = 'Water Relative permeability(krw)',color = 'pink',lw = 5)

plt.xlabel('Water saturation (Fraction)')
plt.ylabel('Relative permeability(fraction)')

plt.grid(True)
plt.legend(loc = 'best')
```

```python
#Step 10: put a title

plt.figure(figsize = (16,10))
plt.plot(sw,kro,label = 'Oil Relative permeability(kro)',color = 'green',lw = 4)
plt.plot(sw,krw,label = 'Water Relative permeability(krw)',color = 'pink',lw = 5)

plt.xlabel('Water saturation (Fraction)')
plt.ylabel('Relative permeability(fraction)')

plt.grid(True)
plt.legend(loc = 'best')
plt.title('Relative Permeability Plot')
```

```python
#step 11: if we want to do scatter ('g--')

plt.figure(figsize = (16,10))
plt.plot(sw,kro,'g--',label = 'Oil Relative permeability(kro)',color = 'green',lw = 4)
plt.plot(sw,krw,'g--',label = 'Water Relative permeability(krw)',color = 'pink',lw = 5)

plt.xlabel('Water saturation (Fraction)')
plt.ylabel('Relative permeability(fraction)')

plt.grid(True)
plt.legend(loc = 'best')
plt.title('Relative Permeability Plot')
```
