## Normal Distribution

> **Syntex: np.random.normal(Loc = mean, scalar = standard deviation, size)**

> To read the diagram properly, follow the below chart

```python
#Example
x = np.random.normal(loc = 5, scale = 5, size = 100000)
```

```python
x
```

```python
se.distplot(x,color = 'green')
```

```python
Z = np.random.normal(loc = 5, scale = 5, size = 100000)
y = np.random.normal(loc = 25, scale = 5, size = 100000)
```

```python
se.distplot(Z,color = 'green')
se.distplot(y,color = 'blue')
plt.grid(True)
```
