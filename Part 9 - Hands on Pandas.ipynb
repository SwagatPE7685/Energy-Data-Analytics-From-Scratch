## 10. Feature Engineering

> Every column in data Terms is called a feature / variable.

> Adding new features to the older data

```python
df.head()
```

```python
df['OIL_VOL_Scaled'] = df['BORE_OIL_VOL']/np.max(df['BORE_OIL_VOL'])
df['WAT_VOL_Scaled'] = df['BORE_WAT_VOL']/np.max(df['BORE_WAT_VOL'])
df['GAS_VOL_Scaled'] = df['BORE_GAS_VOL']/np.max(df['BORE_GAS_VOL'])
```

```python
df.head()
```

```python
#As you can see three column is added

df.shape
```

```python
df[['OIL_VOL_Scaled', 'WAT_VOL_Scaled', 'GAS_VOL_Scaled']].plot(figsize = (12,4), grid = True)
```
